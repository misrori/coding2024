## Example

Find the manufacturer (producer) of the plane that flew the most from JFK Airport in January, you can follow these steps:

````python
# Filter flights departing from JFK in January and count the
tail_number_counts = flights[(flights['origin'] == 'JFK') & (flights['month'] == 1)]['tailnum'].value_counts().reset_index()
tail_number_counts.rename(columns={'count':'number_of_flights'},inplace=True)
#tail_number_counts.columns = ['tailnum', 'n_flights']
tail_number_counts.sort_values('number_of_flights', ascending=False)
tail_number_counts.head(1)
````

````python
(
    flights[(flights["origin"]=='JFK') & (flights['month']==1) ]
    .groupby('tailnum', as_index=False)
    .agg(number_of_flights = ('tailnum', 'count'))
    .sort_values('number_of_flights', ascending=False)
    .head(1)
 )
````

In aggragation you can do the following
*   'count': Count the elements
*   'sum': Sum of the values.
*   'mean': Mean of the values.
*   'median': Median of the values.
*   'min': Minimum value.
*   'max': Maximum value.
*   'std': Standard deviation of the values.
*   'var': Variance of the values.
*   'first': First occurrence.
*   'last': Last occurrence.
*   'nunique': Count of unique values.
