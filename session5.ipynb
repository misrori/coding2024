```python
import pandas as pd

# nycflights13 dataset
# https://cran.r-project.org/web/packages/nycflights13/nycflights13.pdf

flights = pd.read_csv('https://raw.githubusercontent.com/misrori/rdata/main/flights.csv')
airlines = pd.read_csv('https://raw.githubusercontent.com/misrori/rdata/main/airlines.csv')
planes = pd.read_csv('https://raw.githubusercontent.com/misrori/rdata/main/planes.csv')
weather = pd.read_csv('https://raw.githubusercontent.com/misrori/rdata/main/weather.csv')

```


# Filter
**Exercise 1: Filter Late Flights from JFK**

Filter the flights that departed from JFK Airport ('JFK') and had a departure delay greater than 30 minutes. (15241 rows)




```python
df1 = flights[(flights['origin'] == 'JFK') & (flights['dep_delay'] > 30)]
print(df1.shape)
```

**Exercise 2: Filter Long Flights with Arrival Delays**


Filter flights with a distance greater than 1500 miles and an arrival delay greater than 15 minutes.(15553 rows)



```python
df2 = flights[(flights['distance'] > 1500) & (flights['arr_delay'] > 15)]
print(df2.shape)
```


**Exercise 3: Filter American Airlines (AA) Flights with Short Airtime**

Filter American Airlines (carrier code 'AA') flights with an airtime less than 2 hours.(5357 rows)





```python
df3 = flights[(flights['carrier'] == 'AA') & (flights['air_time'] < 120)]
print(df3.shape)
```

**Exercise 4: Filter Early Morning Flights with Long Delays**

Filter flights that departed between 4 AM and 6 AM (hour >= 4 and hour <= 6) and had a departure delay greater than 60 minutes.(602 rows)

```python
df4 = flights[(flights['hour'] >= 4) & (flights['hour'] <= 6) & (flights['dep_delay'] > 60)]
print(df4.shape)
```

## Example

Find the manufacturer (producer) of the plane that flew the most from JFK Airport in January, you can follow these steps:

```python
# Filter flights departing from JFK in January and count the
tail_number_counts = flights[(flights['origin'] == 'JFK') & (flights['month'] == 1)]['tailnum'].value_counts().reset_index()
tail_number_counts.rename(columns={'count':'number_of_flights'},inplace=True)
#tail_number_counts.columns = ['tailnum', 'n_flights']
tail_number_counts.sort_values('number_of_flights', ascending=False)
tail_number_counts.head(1)
```

```python
(
    flights[(flights["origin"]=='JFK') & (flights['month']==1) ]
    .groupby('tailnum', as_index=False)
    .agg(number_of_flights = ('tailnum', 'count'))
    .sort_values('number_of_flights', ascending=False)
    .head(1)
 )
```

In aggragation you can do the following
*   'count': Count the elements
*   'sum': Sum of the values.
*   'mean': Mean of the values.
*   'median': Median of the values.
*   'min': Minimum value.
*   'max': Maximum value.
*   'std': Standard deviation of the values.
*   'var': Variance of the values.
*   'first': First occurrence.
*   'last': Last occurrence.
*   'nunique': Count of unique values.



## Task 1

Calculate the average departure delay by carrier for flights departing from JFK Airport in December.

```python
(
    flights[(flights["origin"]=='JFK') & (flights['month']==12) ]
    .groupby('carrier', as_index=False)
    .agg(average_dep_delay = ('dep_delay', 'mean'))
    .sort_values('average_dep_delay', ascending=False)
)
```

# Task 2
Find the top 10 destinations from JFK Airport in August

```python
(
    flights[(flights["origin"]=='JFK') & (flights['month']==8) ]
    .groupby('dest', as_index=False)
    .agg(number_of_flights = ('dest', 'count'))
    .sort_values('number_of_flights', ascending=False)
)
```

# Task 3
## Most Delayed Flights by Destination


```python
(
    flights[(flights["origin"] == 'JFK')]
    .groupby('dest', as_index=False)
    .agg(
            average_arrival_delay=('arr_delay', 'mean'),
            number_of_flights=('dest', 'count')
        )
    .sort_values('average_arrival_delay', ascending=False).
    head(10)
)
```

# Task 4

##Number of Flights by Carrier in February



```python
(
    flights[flights['month'] == 2]
    .groupby('carrier', as_index=False)
    .agg(number_of_flights=('flight', 'count'))
    .sort_values('number_of_flights', ascending=False)
    .merge(airlines, on='carrier')
    .filter(['number_of_flights', 'name'])
)
```

# Task 5

### The average delays by plane manufacturer where departure delays are grater than 60


```python
(
    flights[flights['dep_delay'] > 60]
    .merge(planes, on='tailnum')
)
```

```python
(
    flights[flights['dep_delay'] > 60]
    .merge(planes, on='tailnum')
    .groupby('manufacturer', as_index=False)
    .agg(average_dep_delay=('dep_delay', 'mean'))
    .sort_values('average_dep_delay', ascending=False)
)
```

#Task 6
Flight count by year of manufacture and origin airport


```python
(
    flights.merge(planes, on='tailnum')
    .groupby(['year_x', 'origin'], as_index=False)
    .agg(number_of_flights=('flight', 'count'))
    .sort_values('number_of_flights', ascending=False)
)
```

#Task 7

Flight Delays by Plane Age (Age = Current Year - Manufacture Year)

after merge, use .assign to create a new variable  like this:

.assign(age=lambda x: 2024 - x['plane_year_column'])  

```python
(
    flights.merge(planes, on='tailnum')
    .assign(age=lambda x: 2024 - x['year_y'])
    .groupby('age', as_index=False)
    .agg(average_arr_delay=('arr_delay', 'mean'),
         number_of_flights=('flight', 'count'))
    .sort_values('average_arr_delay', ascending=False)
)
```

# Task 8
Average departure delay by the Most Common Plane Models from JFK

```python
(
    flights[flights['origin'] == 'JFK']
    .merge(planes, on='tailnum')
    .groupby('model', as_index=False)
    .agg(average_dep_delay=('dep_delay', 'mean'))
    .sort_values('average_dep_delay', ascending=False)
)
```
